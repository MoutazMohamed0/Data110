```python
import matplotlib.pyplot as plt
print(plt.style.available)
```

```python
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.plot(x, x**3, color='green', label='X^3')
plt.plot(x, np.exp(x), color='yellow', label='e^x')
plt.plot(x, np.log(x, where=(x>0)), color='red', label='log(x)')
plt.xlim(0, 5)
plt.ylim(0, 10)

plt.legend()
```

```python
plt.plot(x, x**2, color='g', linestyle='solid')
plt.plot(x, x**3, color='0.75', linestyle='dashed')
plt.plot(x, x**4, color='#FFDD44', linestyle='dashdot')
plt.plot(x, x**5, color='purple', linestyle='dotted')
plt.xlim(0, 1.5)
plt.ylim(0, 5);
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, ((1)/(x**2 + 1)))
plt.xlim(-10, 10)
plt.ylim(-0.5, 1.25)
plt.axis('tight')
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, ((1)/(x**2 + 1)))
plt.xlim(10, -10)
plt.ylim(1.25, -0.5)
plt.axis('equal')
```

```python
plt.plot(x, np.sqrt(x), color='green', label='sqrt(x)')
x_filtered = x[x >= 0]
plt.plot(x, x**(1/3), color='red', label='x^(1/3)')
plt.title("sqrt(x) & x^(1/3)")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = x*np.sin(x)
plt.plot(x, y, '-ok', color='black',
         markersize=7, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.xlim(0, 13)
plt.ylim(-12, 9)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y = x*np.cos(x)
plt.plot(x, y, '-p', color='red', markerfacecolor='gold')
```

```python
plt.style.use('ggplot')
plt.plot(x, np.tan(x), color='#DC143C', label='tan(x)')
plt.plot(x, np.arctan(x), color='#808000', label='arctan(x)')
plt.xlim(-1.5, 1.5)
plt.ylim(-5, 5)
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
```

I chose tan(x) and arctan(x) because the functions are inverses and is it interesting to see their relationship when plotted on a graph. With the x and y limits that I chose you can easily see that the functions intersect at 0. I chose ggplot due to its ease of readability, its dark background makes the graphs more aesthetically pleasing to look at when compared to the default or seaborn-whitegrid styles.

```python
plt.style.use('default')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.style.use('seaborn-whitegrid')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

```python
plt.style.use('ggplot')
x = np.linspace(-2, 4, 1000)
plt.plot(x, x**2 - 2*x + 1)
```

The default style completely lacks a grid and is the simplest of the three styles, it would be best used to display simple datasets. The seaborn-whitegrid style has a light colored background and a grid, and would be best used to display slightly more complex datasets. The ggplot style has a dark colored background and a grid, the default color of the graph also changed from blue to orange, it would best be used for multiple lines and as part of reports or publications as it has the best aesthetics and readability. Because the data not overly complex you could use any style but ggplot is the most aesthetically pleasing.
