```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
y = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5 ]
colors = ['red', 'green', 'yellow', 'blue', 'purple', 'orange', 'gold']
plt.figure(figsize=(10, 10))
plt.bar(x, y, color=colors)

plt.xlabel('Crime')
plt.ylabel('Offenses per 100,000 People')
plt.title('Crime Rates for 2023')
ax = plt.gca()
ax.tick_params(axis='x', labelsize=8)
plt.show()
```

```python
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
y = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
colors = ['red', 'green', 'yellow', 'blue', 'purple', 'orange', 'gold']
plt.barh(x, y, color=colors)

plt.xlabel('Change in Offenses per 100,000 People ')
plt.ylabel('Crime')
plt.title('Changes in Crime Rates from 2022 to 2023')
plt.show()
```

This bar graph makes more sense than task 1's bar graph because it shows a change in data from 2022 to 2023 rather than raw data. Seeing change over time is more insightful than seeing the total number of offenses for a year because it gives you a greater contextual understanding of the trends and past statistics that raw numbers cant capture. It helps us understand that without statistical context data is not as useful. Using these statistics in task 2 we can see that there was an overall decrease in all but one type of crime from 2022 to 2023.

```python
#Changing x ticks, x limit, spine visibility, line style, and adding a horizontal grid behind the bars
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
y = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
colors = ['violet', 'indigo', 'blue', 'green', 'yellow', 'orange', 'red']
plt.barh(x, y, color=colors)
plt.xticks([-12, -8, -4, 0, 4, 8, 12])
plt.xlim(-13, 13)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.grid(axis='y', linestyle='-', color='gray', zorder=0)
plt.xlabel('Change in Offenses per 100,000 People ')
plt.ylabel('Crime')
plt.title('Changes in Crime Rates from 2022 to 2023')
plt.show()
```

Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023:

```python
x = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
y= [264.1/(1-(3.3/100)), 250.7/(1-(8.1/100)), 1347.2/(1-(4.9/100)), 318.7/(1+(12.0/100)), 5.7/(1-(12.0/100)), 38.0/(1-(9.8/100)), 66.5/(1-(0.8/100))]
z = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
x_indexes = np.arange(len(x))
plt.figure(figsize=(11, 11))
ax.tick_params(axis='x', labelsize=5)
print(x_indexes)
plt.bar(x_indexes - 0.125, y, width=0.25, label='2022')
plt.bar(x_indexes + 0.125, z, width=0.25, label='2023')
plt.xlabel('Crime')
plt.ylabel('Offenses per 100,000 People')
plt.title('Crime Rates in 2022 & 2023')
plt.xticks(ticks=x_indexes, labels=x)
plt.legend()
plt.show()
```
